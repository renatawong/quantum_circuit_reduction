```python
from qiskit import QuantumCircuit
from quantum_circuit_reduction import reduce_circuit_depth

'''
Example
'''

qc = QuantumCircuit(2)

qc.h(0)
qc.x(0)
qc.h(0)
qc.x(1)
qc.cx(0, 1)
qc.cx(1, 0)
qc.h(0)
qc.h(0)
qc.sdg(0)
qc.s(0)
qc.y(1)
qc.cx(1,0)
qc.h(0)
qc.draw('mpl')
```

```python
reduced_circuit = reduce_circuit_depth(qc)
reduced_circuit.draw('mpl')
```
